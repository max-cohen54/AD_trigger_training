```python
import numpy as np
import ensembler_functions as ef
import h5py
```

```python
ops_ntuple_file = '/eos/home-m/mmcohen/ad_trigger_development/ops/data/ntuples/data_dict_20250508_482596.h5'
ops_data_dict = {}
with h5py.File(ops_ntuple_file, 'r') as hf:
        for key in hf.keys():
            ops_data_dict[key] = hf[key][:]
```

```python
ef_datasets = ef.load_subdicts_from_h5('/eos/home-m/mmcohen/ad_trigger_development/data/loaded_and_inferenced_data/04-30-2025_regular')

ef_datasets = ef.combine_data(ef_datasets, tags_to_combine=['EB_test', 'EB_train', 'EB_val'], new_tag='EB')
```

```python
print(f'ops data dict:\n')
for key, value in ops_data_dict.items():
    print(f'{key}: {value.shape}')
```

```python
for tag, data_dict in ef_datasets.items():
    print(f'\n{tag}:')
    for key, value in data_dict.items():
        print(f'  {key}: {value.shape}')
```

```python
# Check if L1AD_scores and HLTAD_scores match between ops_data_dict and ef_datasets['EB']

# Extract relevant data from ops_data_dict
ops_event_numbers = ops_data_dict['event_numbers']
ops_run_numbers = ops_data_dict['run_numbers']
ops_L1AD_scores = ops_data_dict['L1AD_scores']
ops_HLTAD_scores = ops_data_dict['HLTAD_scores']

# Extract relevant data from ef_datasets['EB']
ef_event_numbers = ef_datasets['EB']['event_numbers']
ef_run_numbers = ef_datasets['EB']['run_numbers']
ef_L1AD_scores = ef_datasets['EB']['topo2A_AD_scores']
ef_HLTAD_scores = ef_datasets['EB']['HLT_AD_scores']

# Loop over all event number, run number pairs in ops_data_dict
for i in range(len(ops_event_numbers)):
    ops_event = ops_event_numbers[i]
    ops_run = ops_run_numbers[i]

    # Find matching index in ef_datasets['EB']
    match_indices = np.where((ef_event_numbers == ops_event) & (ef_run_numbers == ops_run))[0]

    if len(match_indices) == 0:
        #print(f"No match found for event {ops_event}, run {ops_run} in ef_datasets['EB']")
        continue

    # Assuming one-to-one match, take the first match
    ef_index = match_indices[0]

    # Compare L1AD_scores
    if np.abs(ops_L1AD_scores[i] - ef_L1AD_scores[ef_index]) > 0.01:
        print(f"Mismatch in L1AD_scores for event {ops_event}, run {ops_run}")
        print(f"ops_L1AD_scores[i]: {ops_L1AD_scores[i]}")
        print(f"ef_L1AD_scores[ef_index]: {ef_L1AD_scores[ef_index]}")

    # Compare HLTAD_scores
    if np.abs(ops_HLTAD_scores[i] - ef_HLTAD_scores[ef_index]) > 0.01:
        print(f"Mismatch in HLTAD_scores for event {ops_event}, run {ops_run}")
        print(f"ops_HLTAD_scores[i]: {ops_HLTAD_scores[i]}")
        print(f"ef_HLTAD_scores[ef_index]: {ef_HLTAD_scores[ef_index]}")
```
